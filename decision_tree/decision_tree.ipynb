## Import 

```python
import numpy as np 
import pandas as pd
from graphviz import *
from decision_tree import *
from random_forest import *
```

## Dataset Weather

```python
df = pd.read_csv("data/weather.csv")
df
```

```python
data = df.iloc[:,1:-1]
target = df.iloc[:, -1]

clf = DecisionTree()
clf.fit(data, target)

g = clf.draw_tree("weather")
g
```

```python
print(clf.predict(data))
print(clf.score(data, target))
```
